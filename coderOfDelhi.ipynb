```python
#write a function to display users and their connections
print("\nusers and their connections")
def display_usersdata(data):
    for user in data['users']:
        print(f"ID {user['id']} : {user['name']}")
        print(f"Amit is friends with: {user['friends']} and liked are {user['liked_pages']}")
    print("\npages information")
    for page in data['pages']:
        print(f"{page['id']}:{page['name']}")
display_usersdata(data)
```

```python
#handling missing values
#faulty values etc
import json
def load(name):
    with open(name,'r') as f:
        data2=json.load(f)
        return data2
data2=load('data2.json')
data2
```

```python
def clean_data(data):
    # For cleaning users with empty/whitespace-only names
    data["users"] = [user for user in data['users'] if user['name'].strip()]
    
    # For removing duplicate friends 
    for user in data['users']:
        user['friends'] = list(set(user['friends']))
    
    # Removing users with no friends or likes (inactive)
    data['users'] = [user for user in data['users'] if user['friends'] or user['liked_pages']]
    
    # Removing duplicate pages|
    unique_pages = {}
    for page in data['pages']:
        unique_pages[page['id']] = page
    data['pages'] = list(unique_pages.values())
    return data


# Cleaning the data
data = clean_data(data2)
with open('cleaned_data2.json', 'w') as f:
    json.dump(data, f, indent=4)
data
```

```python
#people you may know feature
def load(data):
    with open(data,'r') as f:
        return json.load(f)
    
def find_people_you_may_know(user_id,data):
    user_friends={}
    for user in data['users']:
        user_friends[user['id']]=set(user['friends'])
    if user['id'] not in user_friends:
        return []
    direct_friends=user_friends[user_id]
    suggestions={}
    for friend in direct_friends:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in direct_friends:
                #Count mutual friends
                suggestions[mutual]=suggestions.get(mutual,0)+1
    sorted_suggestions=sorted(suggestions.items(),key=lambda x:x[1],reverse=True)
    return [user_id for user_id,_ in sorted_suggestions]
#load data
user_id=5
data=load('massive_data.json')
recc=find_people_you_may_know(user_id,data)
print(recc)
```
