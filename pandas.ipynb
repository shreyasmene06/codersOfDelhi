```python
#removing rows to clean empty cells
import pandas as pd
df=pd.read_csv('country.csv')
new_df=df.dropna() #this is if we want to do this in new data frame
print(new_df.to_string())
```

```python
#if we want to modify the existing data frame then we can use inplace
df.dropna(inplace=True)
print(df.to_string())
```

```python
#Method 2
#this is replacing the empty values
df.fillna(130,inplace=True)
```

```python
#for specific rows
df.fillna({"Calories":130},inplace=True)
```

```python
#Mean,Median,Mode
# Since the dataframe only contains text data (country names and codes),
# we cannot calculate mean. Here are some alternatives:

# To get information about the Name column:
print("Number of countries:", len(df["Name"]))

# Or if you want to work with the row containing Afghanistan:
afghanistan_row = df[df["Name"] == "Afghanistan"]
print(afghanistan_row)
```
